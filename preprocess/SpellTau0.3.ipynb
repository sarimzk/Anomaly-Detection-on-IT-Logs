```python
import Spell
import sys
sys.path.append('../')
```

```python
input_dir  = '/Users/vikrant/jupyternotebooks/logs/HDFS/'  # The input directory of log file
output_dir = 'Spell_result/'  # The output directory of parsing results
log_file   = 'named.csv'  # The input log file name
log_format = '<Time> <Machine> <Daemon> <Content>'  # HDFS log format
tau        = 0.3  # Message type threshold (default: 0.5)
regex      = []  # Regular expression list for optional preprocessing (default: [])
```

```python
parser = Spell.LogParser(indir=input_dir,outdir=output_dir, log_format=log_format, tau=tau, rex=regex)
parser.parse(log_file)
```
